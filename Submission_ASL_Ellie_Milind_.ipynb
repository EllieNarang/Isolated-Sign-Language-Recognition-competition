Real-time sign language recognition requires the model to process video data in real-time, which can be computationally intensive. Limited computing resources can lead to slower performance and lower accuracy.

To address the challenge of limited computing resources, several approaches have been proposed. # **Loading and Preprocessing**
The format of the dataset is comparable to the regular MNIST dataset. It contains training and test instances representing labels (0-25), each corresponding to a letter from A to Z. The dataset does not contain examples of the label 9-J and label 25-Z gestures. Each case's format includes a label and a 28x28 pixel image with grayscale values between 0 and 255. This is expressed as the label pixel1, pixel2,..., pixel784 in a header row.

```python
Train_Data=pd.read_csv('/Users/milindpatil/Downloads/sign_mnist_train.csv')
Test_Data=pd.read_csv('/Users/milindpatil/Downloads/sign_mnist_test.csv')
```

```python
Train_Lab=Train_Data['label']
Trainset=Train_Data.drop(['label'],axis=1)
```

The data frame is turned into a NumPy array for DNN training. The array is converted from 1-D to 3-D, which is required for the DNN's first layer. A comparable transformation is performed on the test data frame.

```python
X_Train = Trainset.values
X_Train = Trainset.values.reshape(-1,28,28,1)
print(X_Train.shape)
```

```python
Test_Lab=Test_Data['label']
X_Test=Test_Data.drop(['label'],axis=1)
print(X_Test.shape)
X_Test.head(10)
```

#**Converting the integer labels to binary form**
Each image within the label dataframe is assigned a numeric value ranging from 1 to 24. As the classifier must deal with multiple labels, the DNN output layer contains 24 nodes. Each integer is represented by a 24-by-24 binary array with the corresponding label set to 1 and the other labels set to 0. This procedure employs the LabelBinarizer package from the sklearnpreprocessing library.

```python
from sklearn.preprocessing import LabelBinarizer
Label=LabelBinarizer()
Y_Train=Label.fit_transform(Train_Lab)
Y_Test=Label.fit_transform(Test_Lab)
```

```python
Y_Train
```

```python
X_Test=X_Test.values.reshape(-1,28,28,1)
```

```python
print(X_Train.shape,Y_Train.shape,X_Test.shape,Y_Test.shape)
```

#**Augmentation of the dataset**

The ImageDataGenerator package in keras.preprocessing.image permits the introduction of various distortions into image datasets by randomly rotating, zooming in/out, and scaling the height or width of every pixel. In addition, the dataset is normalized with the rescale parameter, which divides each pixel by 255 to ensure that pixel values range from 0 to 1.

```python
train_datagen = ImageDataGenerator(rescale = 1./255,
                                  rotation_range = 0,
                                  height_shift_range=0.2,
                                  width_shift_range=0.2,
                                  shear_range=0,
                                  zoom_range=0.2,
                                  horizontal_flip=True,
                                  fill_mode='nearest')

X_Test=X_Test/255
```

# **Model Architecture**



```python
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.models import Model

inputs = Input(shape=(28, 28, 1))
x = Conv2D(128, kernel_size=(5,5), strides=1, padding='same', activation='relu')(inputs)
x = MaxPooling2D(pool_size=(3,3), strides=2, padding='same')(x)
x = Conv2D(64, kernel_size=(2,2), strides=1, padding='same', activation='relu')(x)
x = MaxPooling2D(pool_size=(2,2), strides=2, padding='same')(x)
x = Conv2D(32, kernel_size=(2,2), strides=1, padding='same', activation='relu')(x)
x = MaxPooling2D(pool_size=(2,2), strides=2, padding='same')(x)
x = Flatten()(x)
x = Dense(units=512, activation='relu')(x)
x = Dropout(rate=0.25)(x)
outputs = Dense(units=24, activation='softmax')(x)

Model = Model(inputs=inputs, outputs=outputs)
Model.summary()
```

```python

Model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['Accuracy'])
```

#**Training the model**

```python
Model.fit(train_datagen.flow(X_Train,Y_Train,batch_size=200),
         epochs = 40,
          validation_data=(X_Test,Y_Test),
          shuffle=1
         )
```

# **Applying the model and Measuring the Accuracy**

```python
(ls,acc)=Model.evaluate(x=X_Test,y=Y_Test)
```

```python
print('MODEL ACCURACY = {}%'.format(acc*100))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Get predictions for test set
Y_Pred = Model.predict(X_Test)

# Convert predictions to labels
Y_Pred = np.argmax(Y_Pred, axis=1)
Y_True = np.argmax(Y_Test, axis=1)

# Compute confusion matrix
Conf_Mat = confusion_matrix(Y_True, Y_Pred)

# Define class labels
Classes = np.arange(24)

# Plot confusion matrix
fig, ax = Plot.subplots(figsize=(20, 20))
im = ax.imshow(Conf_Mat, cmap='Greens')

# Add colorbar
cbar = ax.figure.colorbar(im, ax=ax)

# Set labels, ticks, and grid
ax.set_xticks(np.arange(len(Classes)))
ax.set_yticks(np.arange(len(Classes)))
ax.set_xticklabels(Classes)
ax.set_yticklabels(Classes)
ax.xaxis.tick_top()
ax.xaxis.set_label_position('top')
ax.set_xlabel('Predicted Label\n')
ax.set_ylabel('True Label\n')
ax.grid(False)

# Add text annotations
Thresh = Conf_Mat.max() / 2.
for i in range(len(Classes)):
    for j in range(len(Classes)):
        ax.text(j, i, format(Conf_Mat[i, j], 'd'),
                ha="center", va="center",
                color="white" if Conf_Mat[i, j] > Thresh else "black")

# Show plot
Plot.tight_layout()
Plot.show()
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Get predictions for test set
Y_Pred = Model.predict(X_Test)

# Convert predictions to labels
Y_Pred = np.argmax(Y_Pred, axis=1)
Y_True = np.argmax(Y_Test, axis=1)

# Compute confusion matrix
Conf_Mat = confusion_matrix(Y_True, Y_Pred)

# Compute percentage of each label in the confusion matrix
Total_True_Labels = np.sum(Conf_Mat, axis=1)
Conf_Mat_Percentage = Conf_Mat / Total_True_Labels[:, np.newaxis]

# Define class labels
Classes = np.arange(24)

# Plot confusion matrix
fig, ax = Plot.subplots(figsize=(20, 20))
im = ax.imshow(Conf_Mat_Percentage, cmap='Purples')

# Add colorbar
cbar = ax.figure.colorbar(im, ax=ax)

# Set labels, ticks, and grid
ax.set_xticks(np.arange(len(Classes)))
ax.set_yticks(np.arange(len(Classes)))
ax.set_xticklabels(Classes)
ax.set_yticklabels(Classes)
ax.xaxis.tick_top()
ax.xaxis.set_label_position('top')
ax.set_xlabel('Predicted Label\n')
ax.set_ylabel('True Label\n')
ax.grid(False)

# Add text annotations
for i in range(len(Classes)):
    for j in range(len(Classes)):
        ax.text(j, i, format(Conf_Mat_Percentage[i, j], '.2%'),
                ha="center", va="center",
                color="white" if Conf_Mat_Percentage[i, j] > 0.5 else "black")

# Show plot
Plot.tight_layout()
Plot.show()
```

# **Model Prediction Using the Camera to Show the Output**

```python
# Capture the webcam input
import cv2
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()

    # Preprocess the input
    input_size = (28, 28)
    preprocessed_frame = cv2.resize(frame, input_size)
    #preprocessed_frame = cv2.resize(frame, input_size, interpolation=cv2.INTER_LINEAR)
    
    preprocessed_frame = cv2.cvtColor(preprocessed_frame, cv2.COLOR_BGR2GRAY)
    preprocessed_frame = np.reshape(preprocessed_frame, (1, *input_size, 1))
    preprocessed_frame = preprocessed_frame / 255.0

    # Apply the model to the input
    prediction = Model.predict(preprocessed_frame)

    # Postprocess the output
    class_index = np.argmax(prediction)
    class_label = chr(ord('A') + class_index)

    # Display the results
    cv2.putText(frame, class_label, (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)
    cv2.imshow('Sign Language Recognition', frame)

    # Exit on key press
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
```
